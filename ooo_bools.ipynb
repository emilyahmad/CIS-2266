```python
try: #validate
    score = float(input("Enter a score between 0.0 and 1.0\n"))
    if (score >= .9):
        print("A")
    elif (score >= .8):
        print("B")
    elif (score >= .7):
        print("C")
    elif (score >= .6):
        print("D")
    else:
        print("F")
except:
    print("Please enter a valid number between 0.0 and 1.0")
```

```python
#don't need validation
hours = int(input("Enter hours worked: "))
rate = float(input("\nEnter hourly rate: "))
if (hours < 40):
    pay = hours * rate
    print("\nTotal pay: " + str(pay))
else:
    base_pay = 40 * rate
    overtime_pay = (hours - 40) * (1.5 * rate)
    print("\nTotal pay: " + "base pay ($" + str(base_pay) + ") + overtime pay ($" + str(overtime_pay) + ") = total pay, $" + str(base_pay + overtime_pay))
```
