## Functions

```python
def computepay(hours, rate):
    if hours <= 40:
        return hours * rate
    else:
        overtime_hours = hours - 40
        return (40 * rate) + (overtime_hours * rate * 1.5)


# Prompt user
hrs_str = input("Enter Hours: ")
rate_str = input("Enter Rate: ")

hrs = float(hrs_str)
rate = float(rate_str)

pay = computepay(hrs, rate)
print(pay)
```

## Lists

```python
filename = "romeo.txt"
words_list = []

with open(filename, "r") as f:
    for line in f:
        words = line.split()
        for w in words:
            if w not in words_list:
                words_list.append(w)

words_list.sort()

# print vertically
for w in words_list:
    print(w)
```

## Dictionaries

```python
filename = "mbox-short.txt"
counts = {}

with open(filename, "r") as f:
    for line in f:
        if line.startswith("From "):
            words = line.split()
            email = words[1]
            counts[email] = counts.get(email, 0) + 1

# Find the email with the greatest count
max_email = None
max_count = None

for email, count in counts.items():
    if max_count is None or count > max_count:
        max_email = email
        max_count = count

print(max_email, max_count)
```
