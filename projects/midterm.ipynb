```python
# Import libraries
import numpy as np
import pandas as pd

# Read in data
ev_data = pd.read_csv("/Users/emilyahmad/Desktop/coding projects/CIS-2266/data/electric_vehicle_data.csv")

# Function definitions

# The identify vehicle function allows users to explore basic
# information about a specific vehicle (user chooses to identify)
def identify_vehicle():
    try:
        choice = int(input("\nIdentify vehicle using: \n1. VIN\n2. DOL Vehicle ID"))
        if (choice == 1):
            user_vehicle = input("\nEnter a Vehicle Identification Number (ex. 5YJ3E1EBXJ, 1N4BZ0CP2G): ")
            chosen_vehicle_data = ev_data[(ev_data["VIN (1-10)"] == user_vehicle)]
            print(chosen_vehicle_data.head())
        elif (choice == 2):
            user_vehicle = int(input("\nEnter a DOI Vehicle ID (ex. 131356970, 211702123): "))
            chosen_vehicle_data = ev_data[(ev_data["DOL Vehicle ID"] == user_vehicle)]
            print(chosen_vehicle_data.head())
    except:
        print("\nPlease enter a valid VIN/DOI")

# The explore geo function filters data based on chosen geographic metric (county, city, etc.)
# and a chosen area (county of Yakima, city of Marysville, etc.) and explore how many
# EVs are registered there or the average make and year of cars registered
def explore_geo():

    option = int(input("\nOptions:\n1. How many cars are registered\n2. Average year of cars registered"))
    
    choice = int(input("\nExplore vehicle info based on geographic data, choose a filter: (use numbers)\n1. County, \n2. City, \n3. State, \n4. Postal Code, \n5. Legislative district"))
    geo_metrics = ["County", "City", "State", "Postal Code", "Legislative District"]
    
    try:
        if (choice == 1):
            specified_area = str(input("\nPick a county (ex. Yakima, King)"))
        elif (choice == 2):
            specified_area = str(input("\nPick a city (ex. Kent, Marysville)"))
        elif (choice == 3):
            specified_area = str(input("\nPick a State (just WA)"))
        elif (choice == 4):
            specified_area = int(input("\nPick a postal code (ex. 98902)"))
        elif (choice == 5):
            specified_area = int(input("\nPick a legislative district (ex. 14)"))
    
    except:
        print("\nPlease choose a valid entry from the list")    
    
    choice = geo_metrics[choice - 1]
    
    if (option == 1):
        print(ev_data[ev_data[choice] == specified_area].shape[0])
    elif (option == 2):
        chosen_geo_area = ev_data[ev_data[choice] == specified_area]
        print("Average make: " + chosen_geo_area["Make"].value_counts().index[0] + ", average year: ")
        print(chosen_geo_area["Model Year"].mean())
    else:
        print("\nPlease choose a valid entry from the list")

# User chooses a make, can explore most popular model and model year, EV type,
# clean alternative fuel vehcile eligibility, electric range, and electric utility
def vehicle_stats():
    # Display car makes in data set (so user knows options)
    make_list = ev_data["Make"].unique()
    index = 1
    for car in ev_data["Make"].unique():
        print(str(index) + ". " + car)
        index += 1

    try:
        make = input("\nChoose a car brand from dataset (case-sensitive): ")
        make_data = ev_data[ev_data["Make"] == make]
        print("\nAverage model year: ")
        print(make_data["Model Year"].mean())

        print("\nMost common model: " + make_data["Model"].value_counts().index[0])

        print("\nEV type: " + make_data["Electric Vehicle Type"].value_counts().index[0])

        print("\nClean alternative fuel vehicle car eligibility: " + make_data["Clean Alternative Fuel Vehicle (CAFV) Eligibility"].value_counts().index[0])

        print("\nElectric range: ")
        print(make_data["Electric Range"].mean())

        print("\nElectric utility: " + make_data["Electric Utility"].value_counts().index[0])

    except:
        print("\nPlease choose a brand from the dataset")

# The explore data set function familiarizes users with the dataset and shows them
# the different functions they can use to filter/continue exploring
def explore_dataset():
    index = 1
    print("\nWhat would you like to learn more about?")
    # loop through column names and add a number next to it
    for col in ev_data.columns:
        print(str(index) + ". " + col)
        index += 1
    choice = int(input("\nSelect an option: "))
    
    # Suggest/limit functions to perform based on data user is interested in
    # Store which function selected
    
    if choice in [1, 13, 14]:
        fx = int(input("\n1. Search for a specific vehicle using VIN or DOL Vehicle ID\n2. Select another column"))
        if (fx == 2):
            explore_dataset()
        else:
            identify_vehicle()
    elif choice in [2, 3, 4, 5, 12]:
        fx = int(input("\n1. Explore geographic vehicle information\n2. Select another column"))
        if (fx == 2):
            explore_dataset()
        else:
            explore_geo()
    elif choice in [6, 7, 8, 9, 10, 11, 15]:
        fx = int(input("\n1. Explore vehicle specific information\n2. Select another column"))
        if (fx == 2):
            explore_dataset()
        else:
            vehicle_stats()
    else:
        explore_dataset()

# Start program
explore_dataset()
```
